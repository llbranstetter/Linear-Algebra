````python
# define a funtion to multiply matricies
def Matrix_Multiplication(M_A, M_B):
    # check the sizes of M_A and M_B
    # print dimentions of A
    print(M_A.shape)
    # print dimentions of B
    print(M_B.shape)
     # create an empty array with the correct dimentions to store the product.
    M_AB = np.zeros((M_A.shape[0], M_B.shape[1]))
    
    # if the collums of A != the rows of B, return an error message
    if (M_A.shape[1] != M_B.shape[0]):
        print("Value is not defined")
        return None
    
    # otherwise proceed with multiplication
    else:
        # for every element in the row
        for a in range(M_AB.shape[0]):
            # and every element in the column
            for b in range(M_AB.shape[1]):
                for c in range(M_A.shape[1]):
                    # compute and insert the element
                    M_AB[a, b] += M_A[a, c] * M_B[c, b]
       
        
    return M_AB 
````

````python
# define a funtion to transpose a matrix
def Transpose(M):
    # create a blank array 
    M_t = np.empty((M.shape[1], M.shape[0]))
    # loop through the entire array and swap the values for the rows and columns
    for i in range(M.shape[0]):
        for j in range(M.shape[1]):
            M_t[j,i] = M[i,j]
    return M_t
tra = Transpose(C)
print(tra)
````
